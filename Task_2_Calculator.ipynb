```python
import tkinter as tk

def click_event(numbers):
    global operator
    operator = operator + str(numbers)
    text.set(operator)
    
def cleartext():
    global operator
    operator=""
    text.set("")
    
def suminput():
    global operator
    equal=str(eval(operator))
    text.set(equal)
    operator=""
    
cal = tk.Tk()
cal.config(bg='salmon')
cal.title("Calculator")
cal.geometry("350x470")
operator=""
text = tk.StringVar()

display = tk.Entry(cal, font=('arial',20,'bold'), textvariable=text, justify="right", bd=30, insertwidth=4, bg='salmon')
display.grid(columnspan=3)

btnadd = tk.Button(cal,fg='black', font=(20), text="+", width=5, height=2, bd=8, command=lambda:click_event("+"), bg='red')
btnadd.grid(row=1,column=0)

btnsub = tk.Button(cal,fg='black', font=(20), text="-", width=5, height=2, bd=8, command=lambda:click_event("-"), bg='red')
btnsub.grid(row=1,column=1)

btnmul = tk.Button(cal,fg='black', font=(20), text="X", width=5, height=2, bd=8, command=lambda:click_event("*"), bg='red')
btnmul.grid(row=1,column=2)

btndiv = tk.Button(cal,fg='black', font=(20), text="/", width=5, height=2, bd=8, command=lambda:click_event("/"), bg='coral')
btndiv.grid(row=2,column=0)

btnequal = tk.Button(cal,fg='black', font=(20), text="=", width=5, height=2, bd=8, command=suminput, bg='coral')
btnequal.grid(row=2,column=1)

btnclear = tk.Button(cal,fg='black', font=(20), text="C", width=5, height=2, bd=8, command=cleartext, bg='coral')
btnclear.grid(row=2,column=2)

btn7 = tk.Button(cal,fg='black', font=(20), text="7", width=5, height=2, bd=8, command=lambda:click_event(7), bg='red')
btn7.grid(row=3,column=0)

btn8 = tk.Button(cal,fg='black', font=(20), text="8", width=5, height=2, bd=8, command=lambda:click_event(8), bg='red')
btn8.grid(row=3,column=1)

btn9 = tk.Button(cal,fg='black', font=(20), text="9", width=5, height=2, bd=8, command=lambda:click_event(9), bg='red')
btn9.grid(row=3,column=2)

btn4 = tk.Button(cal,fg='black', font=(20), text="4", width=5, height=2, bd=8, command=lambda:click_event(4), bg='coral')
btn4.grid(row=4,column=0)

btn5 = tk.Button(cal,fg='black', font=(20), text="5", width=5, height=2, bd=8, command=lambda:click_event(5), bg='coral')
btn5.grid(row=4,column=1)

btn6 = tk.Button(cal,fg='black', font=(20), text="6", width=5, height=2, bd=8, command=lambda:click_event(6), bg='coral')
btn6.grid(row=4,column=2)

btn1 = tk.Button(cal,fg='black', font=(20), text="1", width=5, height=2, bd=8, command=lambda:click_event(1), bg='red')
btn1.grid(row=5,column=0)

btn2 = tk.Button(cal,fg='black', font=(20), text="2", width=5, height=2, bd=8, command=lambda:click_event(2), bg='red')
btn2.grid(row=5,column=1)

btn3 = tk.Button(cal,fg='black', font=(20), text="3", width=5, height=2, bd=8, command=lambda:click_event(3), bg='red')
btn3.grid(row=5,column=2)

btnzero = tk.Button(cal,fg='black', font=(20), text="0", width=5, height=2, bd=8, command=lambda:click_event(0), bg='coral')
btnzero.grid(row=6,column=0)

btn_zero = tk.Button(cal,fg='black', font=(20), text="00", width=5, height=2, bd=8, command=lambda:click_event(00), bg='coral')
btn_zero.grid(row=6,column=1)

btndot = tk.Button(cal,fg='black', font=(20), text=".", width=5, height=2, bd=8, command=lambda:click_event("."), bg='coral')
btndot.grid(row=6,column=2)

cal.mainloop()
```
