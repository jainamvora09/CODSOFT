```python
import random
import string
from tkinter import *

root =Tk()
#root.config(bg='grey')
root.title("Password Generator")
root.geometry("450x350")


label = Label(root, text="Choose the strength of password", fg='white', bg='black', font=('times new roman',15,'bold'))
label.pack()

def selection():
    selection = choice.get()

choice = IntVar()
weakrb = Radiobutton(root, text="Weak Password", variable=choice, value=1,font=('bold'), command=selection)
weakrb.pack()

mediumrb = Radiobutton(root, text="Medium Password", variable=choice, value=2,font=('bold'), command=selection)
mediumrb.pack()

strongrb = Radiobutton(root, text="Strong Password", variable=choice, value=3,font=('bold'), command=selection)
strongrb.pack()

space = Label(root)
space.pack()

label1 = Label(root, text="Select the length of password", font=('bold'))
label1.pack()

val = IntVar()
box = Spinbox(root, from_=8, to_=20, textvariable=val, width=20)
box.pack()

def callback():
    result.config(text=gPass())

submit = Button(root, text="Generate Password", font=('bold'), command=callback)
submit.pack(pady=10)

result = Message(root, text="", relief=RAISED, width=200, font=(25))
result.pack(side=BOTTOM)

weak = string.ascii_lowercase + string.ascii_uppercase
medium = string.ascii_lowercase + string.ascii_uppercase + string.digits
strong = string.ascii_lowercase + string.ascii_uppercase + string.digits + string.punctuation

def gPass():
    if choice.get() == 1:
        return "".join(random.sample(weak, val.get()))
    elif choice.get() == 2:
        return "".join(random.sample(medium, val.get()))
    elif choice.get() == 3:
        return "".join(random.sample(strong, val.get()))

root.mainloop()
```
